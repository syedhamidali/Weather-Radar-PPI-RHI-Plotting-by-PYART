```python
%%time
import warnings
warnings.filterwarnings("ignore")
import pyart
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
radar = pyart.io.read("../IMD_Radar_stuff/PyScanCf/Examples/goa16/polar_GOA210516024101.nc")
```

```python
pseudorhi = pyart.util.cross_section_ppi(radar, [270])
display = pyart.graph.RadarDisplay(pseudorhi)
fig, ax = plt.subplots(2,1, sharex=True,sharey=True, figsize= (10,10))
display.plot_rhi('VELH', ax = ax[0], vmin = -30, vmax = 30, cmap = "pyart_NWSVel")
display.plot_rhi('REF', ax = ax[1], cmap = "pyart_NWSRef")
ax[0].set_ylim([0,20])
ax[0].set_xlim([0,100])
```
